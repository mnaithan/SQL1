```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import sqlite3 as db
```

```python
from pandasql import sqldf
```

```python
#url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
adult = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",sep=",",header=None)
adult.head()
```

```python
columns = ['age','workclass','fnlwgt','education','education-num','marital_status','occupation','relationship','race','sex','capital_gain','capital_loss',
           'hrs_per_wee','native_country','class']
```

```python
columns
```

```python
adult.columns = columns
```

```python
adult.head()
```

```python
from sqlalchemy import create_engine
disk_engine = create_engine('sqlite:///adult.db')
adult.to_sql('adult', disk_engine, if_exists='append')
```

```python
df = pd.read_sql_query('SELECT * FROM adult LIMIT 10',disk_engine)
len(df)
```

```python
pd.read_sql_query('SELECT COUNT(DISTINCT occupation) FROM adult',disk_engine)
```
